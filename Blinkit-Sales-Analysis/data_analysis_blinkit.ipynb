#### **Data Cleaning**

```python
print(df["Item Fat Content"].unique())
```

```python
df['Item Fat Content'] = df['Item Fat Content'].replace({'LF': 'Low Fat',
                                                         'low fat': 'Low Fat', 
                                                         'reg': 'Regular'
                                                         })
print(df["Item Fat Content"].unique())
```

#### **Business Requirement AND KPIs (Key Performance Index)**

```python
#  Total Sales
total_sales = df['Sales'].sum()


# Average Sales 
average_sales = df['Sales'].mean()


# Number of Items Sold
total_items_sold = df['Sales'].count()


# Average Ratings
average_rating = df['Rating'].mean()


# Displaying the output
# Keep this in Mind that don't add a {'space'} in between "," and ".1f" otherwise it won't run
print(f"Total Sales : ₹ {total_sales:,.2f}")
print(f"Average Sales : ₹ {average_sales:,.2f}")
print(f"Number of Items Sold : {total_items_sold:,.0f}")
print(f"Average Rating : {average_rating:,.1f}")
```

## **Data Visualization**

#### **Total Sales by Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()
# print(round(sales_by_fat,2))
plt.pie(sales_by_fat, autopct="%1.1f%%", labels=sales_by_fat.index, startangle=90)
plt.title("Sales by Fat Content")
plt.show()

# sales_by_fat.index is used to write the name of fat content on the pie chart 
```

#### **Total Sales by Item Type**

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
print(round(sales_by_type,2))

plt.figure(figsize=(10,6))
bar_plot = plt.bar(sales_by_type.index, sales_by_type)

plt.xticks(rotation=90)
plt.xlabel("Item Type")
plt.ylabel("Total Sales")
plt.title("Sales by Item Type")

# This is just for the numbers that are written on each bars
for bar in bar_plot :
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

#### **Fat Content by Outlet for Total Sales**

```python
fat_content_by_outlet = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
fat_content_by_outlet = fat_content_by_outlet[['Regular', 'Low Fat']]

our_plot = fat_content_by_outlet.plot(kind='bar', figsize=(8,5), title = "Outlet Tier by Item Fat Content")
plt.xlabel("Outlet Location Tier")
plt.ylabel("Total Sales")
plt.legend(title = "Item Fat Content")
plt.tight_layout()
plt.show()
```

#### **Total Sales by Outlet Establishment**

```python
sales_by_year = df.groupby(['Outlet Establishment Year'])['Sales'].sum().sort_index()
print(round(sales_by_year,2))
```

```python
plt.figure(figsize=(9,5))
plt.plot(sales_by_year.index, sales_by_year.values, marker=".", linestyle="-")

plt.xlabel("Outlet Establishment Year")
plt.ylabel("Total Sales")
plt.title("Outlet Establishment")

for x, y in zip(sales_by_year.index, sales_by_year.values) :
    plt.text(x, y, f'{y:,.0f}',ha='center', va="bottom", fontsize=8)

plt.tight_layout()
plt.show()
```

#### **Total Sales by Outlet Size**

```python
sales_by_outlet_size = df.groupby(['Outlet Size'])['Sales'].sum().sort_values(ascending=False)
print(round(sales_by_outlet_size, 2))
```

```python
plt.figure(figsize=(5,5))
plt.pie(sales_by_outlet_size, autopct='%1.1f%%',labels=sales_by_outlet_size.index, startangle=90, explode=[0.02,0,0])
plt.title("Outlet Size")
plt.tight_layout()
plt.show()
```

#### **Sales by Outlet Location**

```python
sales_by_location = df.groupby(['Outlet Location Type'])['Sales'].sum().reset_index()
print(round(sales_by_location,2))
```

```python
plt.figure(figsize=(8,3)) # Smaller Height and More/Enough Width
ax = sns.barplot(x='Sales', y='Outlet Location Type', data=sales_by_location)

plt.title("Total Sales by Outlet Location Type")
plt.xlabel("Total Sales")
plt.ylabel("Outlet Location Type")

plt.tight_layout() # To ensure the Layout fits without Scroll
plt.show()
```
